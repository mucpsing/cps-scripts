```python
from main import Matrix
from PIL import Image
```

```python
# 要处理的文件
filename = r'./test/test.png'

# 输入需要偏移的坐标
xy={ 'left_top':[50,150], 'right_down':[-99, 250] }

# 坐标格式可以可以使用绝对定位和相对定位，使用相对定位演示 relative | absolute
mode = 'relative'

# 执行
M = Matrix(filename).config(mode=mode, xy=xy).draw()
print(M)
```

```python
# 通过 result() 返回 PIL 的image 对象
res = M.result()
print(res)
```
